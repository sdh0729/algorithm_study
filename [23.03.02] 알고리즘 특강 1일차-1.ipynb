```python
# 전역 변수부 
katok = ['다현','정연','쯔위','사나','지효']

# 메인 코드부
print(katok)
```

```python
# (1) 데이터 추가 : 모모가 1회 카톡
katok.append(None)
katok[5] = '모모'
print(katok)
```

```python
# (2) 데이터 삽입 : 미나가 연속으로 40회 카톡
# 1단계 : 빈칸을 준비
katok.append(None)
# 2단계 : 3등까지 친구들을 오른쪽으로 한칸 이동시키기<반복>
katok[6] = katok[5]
katok[5] = None
katok[5] = katok[4]
katok[4] = None
katok[4] = katok[3]
katok[3] = None
# 3단계 : 빈칸(3등) 자리에 미나 삽입
katok[3] = '미나'
print(katok)
```

```python
# (3) 데이터 삭제 : 사나가 카톡을 차단함 == 더이상 친구 아님
# 1단계 : 데이터 제거
katok[4] = None
# 2단계 : 친구들 한명씩 승급 시키기.(마지막 친구까지..)
katok[4] = katok[5]
katok[5] = None
katok[5] = katok[6]
katok[6] = None
# 3단계 : 빈칸 제거
del(katok[6])
print(katok)
```
