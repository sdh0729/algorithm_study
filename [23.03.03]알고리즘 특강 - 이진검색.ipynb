```python
# 함수
import random
def binSearch(ary,fData) :
  pos = -1
  start = 0
  end = len(ary)-1
  while (start <= end) :
    mid = (start + end) // 2
    if (ary[mid] == fData) :
      pos = mid
      break
    elif (ary[mid]< fData) :
      start = mid + 1
    else :
      end = mid - 1

  return pos 

# 전역
dataAry = [random.randint(50, 190) for _ in range(8)]
findData = random.choice(dataAry)
dataAry.sort()
findData = 110

# 메인
print('배열-->', dataAry)
position = binSearch(dataAry, findData)
if (position == -1) :
    print(findData, '못찾음...')
else :
    print(findData, '는', position, '위치에 있어요.')
```
