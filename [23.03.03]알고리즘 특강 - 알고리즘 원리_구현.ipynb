# 선택 정렬


```python
# 함수
import random
def findMinIndex(ary) :
  minIdx = 0
  for i in range(1,len(ary)) :
    if (ary[minIdx] > ary[i]) :
      minIdx = i
  return minIdx
  
# 전역
before = [random.randint(50,190) for _ in range(8)]
after = []
# 메인
print('정렬 전 -->', before)
for _ in range(len(before)):
  minPos = findMinIndex(before)
  after.append(before[minPos])
  del (before[minPos])


print('정렬 후 -->', after)
```

# 선택 정렬 알고리즘 : 사이클 1


```python
# 함수
import random
def selectionSort(ary) :
  n = len(ary)
  for i in range(0,n-1,1) :
    minIdx = i
    for k in range(i+1,n,1) :
      if (ary[minIdx] > ary[k]) :
        minIdx = k
    ary[minIdx],ary[i] = ary[i],ary[minIdx]
  return ary

# 전역
dataAry = [random.randint(50,190) for _ in range(18)]

# 메인

print('정렬 전 -->', dataAry)
dataAry = selectionSort(dataAry)

print('정렬 후 -->', dataAry)
```
