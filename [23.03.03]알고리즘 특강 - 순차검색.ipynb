```python
## 함수
import random
def seqSearch(ary, fData) :
    pos = -1
    for i in range(0, len(ary)) :
        if (ary[i] == fData) :
            pos = i
            break
    return pos

## 전역
dataAry = [random.randint(50, 190) for _ in range(8)]
findData = random.choice(dataAry)

## 메인
print('배열-->', dataAry)
position = seqSearch(dataAry, findData)
if (position == -1) :
    print(findData, '못찾음...')
else :
    print(findData, '는', position, '위치에 있어요.')
```
